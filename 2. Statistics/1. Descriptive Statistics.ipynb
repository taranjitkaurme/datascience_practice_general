## Descriptive Statistics
###  Measures of Central Tendency

Mean, median, and mode are measures of central tendency, which give us an idea of where the center of a distribution or a set of values is.

**Mean**: The mean is simply the average of all the values in a dataset. You calculate it by adding up all the values and then dividing by the number of values.
* Formula: Mean = (Sum of all values) / (Number of values)
* Example: Let's say we have a set of test scores: 85, 90, 92, 88, and 95. The mean would be (85 + 90 + 92 + 88 + 95) / 5 = 90.
* When to use: Use the mean when you want a balance of all values and there are no extreme outliers. It's sensitive to extreme values, so if there are outliers, the mean might not represent the "typical" value well.

**Median**: The median is the middle value in a dataset when it's ordered. If there's an even number of values, the median is the average of the two middle values.
* To find the median, you first need to arrange the data in numerical order.
* Example: Using the same test scores, when ordered: 85, 88, 90, 92, 95. The median is 90, as it's the middle value.
* When to use: Use the median when your data has outliers or is not symmetrically distributed. It's less sensitive to extreme values and gives you a better sense of the central location if there are outliers.

**Mode**: The mode is the value that appears most frequently in a dataset. A dataset can have no mode, one mode, or more than one mode.
* A set of data may have one mode, more than one mode (multimodal), or no mode at all if no value is repeated.
* Example: In the test scores, there is no mode because no score repeats.
* When to use: Use the mode when you want to identify the most common value in your dataset. It's especially useful for categorical data.
 

In summary, the choice between mean, median, and mode depends on the nature of your data and the insights you're seeking.
Note: There can be more than one mode for a given variable dataset. For instance, there can be elections in which three parties participate, two of those get 40% of the votes, each, and the third party gets 20%. In this case, there are two modes since two parties have the highest (equal) number of votes.


When the mean=median=mode the data is said to be normally distributed. When the sample or population distribution is not normal, the central limit theorem states that if you take infinitely many samples from the population, evaluate the means of these samples, the resulting distribution will be a normal distribution.


However, there are some less-known facts and nuances about each of these measures of central tendency:

*Mean (Arithmetic Average)*

* Sensitive to Outliers: The mean is affected by extreme values or outliers. For instance, in a set of incomes, a few very high incomes can significantly increase the mean, making it not truly representative of the general population.

* Not Always a Data Point: In many data sets, especially when dealing with fractional or decimal numbers, the mean might not correspond to an actual data point. For example, the mean of 1, 2, and 4 is 2.33, which is not a value in the dataset.

* Different Types of Means: Apart from the arithmetic mean, there are other types like geometric mean and harmonic mean, each useful in different scenarios. For example, the geometric mean is more appropriate for data that grows exponentially.

*Median (Middle Value)*

* Unaffected by Outliers: Unlike the mean, the median is not influenced by extremely high or low values. This makes it a better measure of central tendency for skewed distributions.

* Better for Ordinal Data: The median can be used for ordinal data (data that can be ranked but not quantitatively measured), whereas the mean requires interval or ratio level data.

* Finding the Median in Even Sets: When there is an even number of data points, the median is the mean of the two middle numbers. This interpolated value might not be an actual data point in the set.

*Mode (Most Frequent Value)*

* Can Be More than One: A data set can have more than one mode (bimodal, trimodal, etc.) if multiple values are equally frequent. This is particularly useful in understanding the distribution of categorical data.

* Useful in Non-Numerical Data: Mode can be applied to non-numerical data, such as survey responses or most common category, which is not possible with mean or median.
Not Always Present: Some data sets may not have a mode at all if no number repeats or if all numbers repeat with the same frequency.

Each of these measures provides different insights and is useful in different scenarios, highlighting the importance of understanding the nature of your data before choosing the appropriate measure of central tendency.

* No Clear Central Tendency: In some cases, none of these measures gives a clear picture of the central tendency. For example, if everyone in your group slept for vastly different amounts of time (like 3 hours, 6 hours, 8 hours, 12 hours, and 15 hours), the mean, median, and mode might not tell you much about a "typical" sleep pattern.

These examples show why it's important to consider the context and the specific data set when choosing and interpreting measures of central tendency. Each measure provides a different type of information, and sometimes you need to use more than one to really understand your data.

```python
import statistics
from scipy import stats

# Example data
test_scores = [85, 90, 92, 88, 95]

# Mean
mean_score = statistics.mean(test_scores)
print(f"Mean: {mean_score}")

# Median
median_score = statistics.median(test_scores)
print(f"Median: {median_score}")

# Mode
# The mode function in scipy returns a mode object, which contains the mode(s) and their frequencies
mode_result = stats.mode(test_scores)
print(f"Mode: {mode_result.mode}, Frequency: {mode_result.count}")
```

#### The five-number summary
The five-number summary is a set of descriptive statistics that provides a concise summary of the distribution of a dataset. It includes the following five values:

1. **Minimum**: The smallest value in the dataset.
2. **First Quartile (Q1)**: The median of the lower half of the dataset. It's the value below which 25% of the data falls.
3. **Median (Second Quartile, Q2)**: The middle value of the dataset, separating it into two equal halves.
4. **Third Quartile (Q3)**: The median of the upper half of the dataset. It's the value below which 75% of the data falls.
5. **Maximum**: The largest value in the dataset.

The five-number summary is particularly useful for describing the spread and central tendency of a dataset, especially when the dataset is large and diverse.

Here's a simple example in Python using the same test scores data:

```python
import numpy as np

# Example data
test_scores = [85, 90, 92, 88, 95]

# Five-number summary using numpy
minimum = np.min(test_scores)
q1 = np.percentile(test_scores, 25)
median = np.median(test_scores)
q3 = np.percentile(test_scores, 75)
maximum = np.max(test_scores)

print(f"Minimum: {minimum}")
print(f"Q1: {q1}")
print(f"Median: {median}")
print(f"Q3: {q3}")
print(f"Maximum: {maximum}")
```

 library:

Another visual way to represent the five-number summary and additional information about the distribution of a dataset is by using a box plot (also known as a box-and-whisker plot). A box plot visually displays the minimum, first quartile (Q1), median, third quartile (Q3), and maximum of a dataset.

In a box plot:

* A box is drawn from Q1 to Q3, representing the interquartile range (IQR).
* A line (whisker) extends from the box to the minimum and maximum values.
* A line inside the box represents the median.
* Outliers, if any, are often shown as individual points beyond the whiskers.

Here's an example in Python using the matplotlib library:

```python
import matplotlib.pyplot as plt
import numpy as np

# Example data
test_scores = [85, 90, 92, 88, 95]

# Create a box plot
plt.boxplot(test_scores)
plt.title('Box Plot of Test Scores')
plt.ylabel('Scores')
plt.show()
```

The next set of descriptive statistics is, the measures of dispersion

Measures of dispersion provide information about how spread out or dispersed the values in a dataset are. There are several measures of dispersion, and I'll explain three common ones: range, variance, and standard deviation.

**Range**: The range is the simplest measure of dispersion and is calculated as the difference between the maximum and minimum values in a dataset. It gives you an idea of how spread out the scores are, but it doesn't tell you much about the scores in between.
* Range=Maximum Value−Minimum Value
* When to use: The range is a quick and easy measure, but it's sensitive to outliers. Use it when you want a simple indication of the spread, especially in datasets without extreme values.
* Example: If you have a set of test scores ranging from 70 to 95, the range is 95 - 70 = 25.

**Variance**: Variance measures how far each data point in the set is from the mean. It is calculated by taking the average of the squared differences between each data point and the mean. It's like asking, "On average, how much is each score different from the average?" Squaring the differences emphasizes bigger deviations.

* When to use: Use variance when you want to consider the squared deviations of each data point from the mean. However, because it involves squared differences, it can be influenced by outliers
* Example: If most scores are close to the average, the variance is low. If they're spread out, the variance is high..

**Standard Deviation**: The standard deviation is the square root of the variance. It provides a more interpretable measure of the spread of data, as it is in the same units as the original data It's a more "friendly" version of variance. It tells you, on average, how much each score differs from the mean, and it's in the same units as your data..
* When to use: Standard deviation is often preferred over variance because it is in the same units as the original data. It gives a more interpretable measure of how spread out the values are. Use standard deviation when you want a more intuitive understanding of the dispersion
* Example: If the standard deviation is 2, it means most scores are within 2 points of the average..

In Python, you can calculate these measures using the numpy library for the variance and standard deviation. Here's an example:

```python
import numpy as np

# Example data
test_scores = [85, 90, 92, 88, 95]

# Range
data_range = np.max(test_scores) - np.min(test_scores)
print(f"Range: {data_range}")

# Variance
variance = np.var(test_scores)
print(f"Variance: {variance}")

# Standard Deviation
std_deviation = np.std(test_scores)
print(f"Standard Deviation: {std_deviation}")
```

So, in essence:

* Range: How spread out are the extreme values?
* Variance: How much, on average, does each score differ from the mean? (with an emphasis on bigger differences)
* Standard Deviation: Like variance, but in a more understandable unit. It tells you how much, on average, each score differs from the mean.

### Measures of Shape

Measures of shape describe the distribution or pattern of values in a dataset. They help us understand the symmetry, skewness, and peakedness of the distribution. Three common measures of shape are skewness, kurtosis, and histograms.

1. **Skewness**:

* What it is: Skewness measures the asymmetry of the distribution.
* Positive Skewness: The distribution is skewed to the right, meaning it has a long tail on the right side. The mean is typically greater than the median.
* Negative Skewness: The distribution is skewed to the left, with a long tail on the left side. The mean is typically less than the median.
* When to use: Skewness helps identify whether the bulk of the data is on one side of the distribution.
*  https://www.scaler.com/topics/data-science/descriptive-statistics/

* Mean < Median : negative, or left-skewed
* Mean = Median : symmetric, or zero-skeweness
* Mean > Median : positive, or right-skewed

2. **Kurtosis**:

* What it is: Kurtosis measures the "tailedness" of the distribution.
* Leptokurtic (Positive Kurtosis): The distribution has heavy tails and a high peak. It indicates more outliers and a higher likelihood of extreme values.
* Platykurtic (Negative Kurtosis): The distribution has lighter tails and a flatter peak. It indicates fewer outliers and less likelihood of extreme values.
* Mesokurtic (Normal Kurtosis): Similar to a normal distribution.
* When to use: Kurtosis helps identify the shape of the tails and the likelihood of extreme values.
* 

3. **Histograms**:

* What it is: A histogram is a visual representation of the distribution of a dataset. It displays the frequency of different ranges or bins of values.
* When to use: Histograms are useful for getting an overall sense of the distribution, including its shape, central tendency, and spread.
  
In Python, you can use libraries like matplotlib to create histograms and scipy or numpy to calculate skewness and kurtosis. Here's a simple example:

```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import skew, kurtosis

# Example data
data = np.random.normal(0, 1, 1000)  # Generating 1000 random samples from a normal distribution

# Plot histogram
plt.hist(data, bins=30, edgecolor='black')
plt.title('Histogram of Random Data')
plt.xlabel('Values')
plt.ylabel('Frequency')
plt.show()

# Calculate skewness and kurtosis
skewness_value = skew(data)
kurtosis_value = kurtosis(data)

print(f"Skewness: {skewness_value}")
print(f"Kurtosis: {kurtosis_value}")
```

* Skewness: 0.1104

**Interpretation**: A positive skewness indicates that the distribution of your data is slightly skewed to the right. This means that there might be a few more data points on the left side of the distribution, and the tail on the right side is longer. The mean is likely to be slightly greater than the median.

* Kurtosis: 0.0425

**Interpretation**: The kurtosis value of 0.0425 is close to zero, suggesting that the distribution has a relatively normal shape (mesokurtic). It doesn't exhibit particularly heavy tails (leptokurtic) or light tails (platykurtic). The peak of the distribution is moderately sharp.
In summary, based on these values, your data seems to have a slightly right-skewed distribution with a relatively normal shape and a moderate peak. Keep in mind that interpretations can vary, and the context of your specific data and analysis goals is crucial for a complete understanding.

The **describe** function in Python, particularly in the pandas library, provides a comprehensive summary of the central tendency, dispersion, and shape of a dataset's distribution. When applied to a DataFrame, it gives various summary statistics for each numerical column. The describe function is a convenient way to quickly analyze and understand the distribution of your data.
The output will include:

count: Number of non-null observations.
mean: Mean (average) of the values.
std: Standard deviation, a measure of the amount of variation or dispersion.
min: Minimum value.
25%: First quartile (Q1).
50%: Median (Q2).
75%: Third quartile (Q3).
max: Maximum value.

Here's an example using pandas:

```python
import pandas as pd

# Example data
test_scores = [85, 90, 92, 88, 95]

# Create a DataFrame
df = pd.DataFrame({'Test Scores': test_scores})

# Use describe to get summary statistics
summary_stats = df.describe()

print(summary_stats)
```

* If the mean=median=mode, then the data is symmetrical data, standard deviation and variance are near to zero or very low, kurtosis is low or near to zero. This means that the data is distributed normally. 
* If mean>median>mode, then the data is skewed positively. Conversely, if the mean<median<mode, then the data is skewed negatively. The standard deviation of such data is non-zero. The higher the difference between the values of mean, median and mode of the data, the higher should be the values of standard deviation and variance of that data. The kurtosis score of such data will be non-zero depending on the peaked-ness of the distribution. All these measures will indicate that the data is not distributed normally.
